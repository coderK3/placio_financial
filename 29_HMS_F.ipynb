```python
import networkx as nx #Graphical persentation of unshorted data
import numpy as np #calculation lib
import pandas as pd #data Shoting Lib eg:series
import matplotlib.pyplot as mlt #for plot the data 
mlt.style.use('fivethirtyeight')
import seaborn as sns #statical data vertualisation 
import plotly.offline as py #plotly is a data_virtualisation 
py.init_notebook_mode(connected=True)
import plotly.graph_objs as go
import plotly.tools as tls

#check_out and ls will help in the to show files and folders

from subprocess import check_output
print(check_output(["ls"]).decode("utf8"))
```

```python
HMS29_ = pd.read_csv('csv/29_july.csv')
HMS29_
```

```python
D29_ = HMS29_.sort_values(by='Collected Amount' , ascending= False)
D29_
```

```python
mlt.subplots(figsize=(10,7))
sns.countplot(x=D29_['Mode'],hue=D29_['Mode'], data=D29_).set_title("Tenant count Payment Mode wise")
#mlt.savefig('status_20180701_to_20180723.png')
mlt.show()
```

```python
mlt.subplots(figsize=(20,9))
sns.countplot(x=D29_['Property Name'],hue=D29_['Property Name'], data=D29_).set_title("Tenant count property wise")
#mlt.savefig('status_20180701_to_20180723.png')
mlt.show()
```

```python
mlt.subplots(figsize=(10,7))
sns.countplot(x=D29_['Mode'],hue=D29_['Term'], data=D29_).set_title("monthly and yearly count payment mode wise")
#mlt.savefig('status_20180701_to_20180723.png')
mlt.show()
```

```python
mlt.subplots(figsize=(20,9))
sns.countplot(x=D29_['Property Name'],hue=D29_['Collected Amount'], data=D29_).set_title("amount collected property wise")
#mlt.savefig('status_20180701_to_20180723.png')
mlt.show()
```

```python
mlt.subplots(figsize=(10,9))
sns.countplot(x=D29_['Property Name'],hue=D29_['Status'], data=D29_).set_title("status Property wise")
#mlt.savefig('status_20180701_to_20180723.png')
mlt.show()
```

```python
mlt.subplots(figsize=(10,7))
sns.countplot(x=D29_['Tenant Code'],hue=D29_['Mode'], data=D29_).set_title("Payment Mode used buy tenant")
#mlt.savefig('status_20180701_to_20180723.png')
mlt.show()
```

```python
mlt.subplots(figsize=(20,7))
sns.countplot(x=D29_['Tenant Code'],hue=D29_['Term'], data=D29_).set_title("Tenant term used")
#mlt.savefig('status_20180701_to_20180723.png')
mlt.show()
```
