```python
import networkx as nx #Graphical persentation of unshorted data
import numpy as np #calculation lib
import pandas as pd #data Shoting Lib eg:series
import matplotlib.pyplot as mlt #for plot the data 
mlt.style.use('fivethirtyeight')
import seaborn as sns #statical data vertualisation 
import plotly.offline as py #plotly is a data_virtualisation 
py.init_notebook_mode(connected=True)
import plotly.graph_objs as go
import plotly.tools as tls

#check_out and ls will help in the to show files and folders

from subprocess import check_output
print(check_output(["ls"]).decode("utf8"))
```

```python
HMS28_ = pd.read_csv('csv/27_july.csv')
HMS28_
```

```python
D28_ = HMS28_.sort_values(by='Collected Amount' , ascending= False)
D28_
```

```python
mlt.subplots(figsize=(10,7))
sns.countplot(x=D28_['Mode'],hue=D28_['Mode'], data=D28_).set_title("Tenant count Payment Mode wise")
#mlt.savefig('status_20180701_to_20180723.png')
mlt.show()
```

```python
mlt.subplots(figsize=(20,9))
sns.countplot(x=D28_['Property Name'],hue=D28_['Property Name'], data=D28_).set_title("Tenant count property wise")
#mlt.savefig('status_20180701_to_20180723.png')
mlt.show()
```

```python
mlt.subplots(figsize=(20,7))
sns.countplot(x=D28_['Mode'],hue=D28_['Term'], data=D28_).set_title("monthly and yearly count payment mode wise")
#mlt.savefig('status_20180701_to_20180723.png')
mlt.show()
```

```python
mlt.subplots(figsize=(20,9))
sns.countplot(x=D28_['Property Name'],hue=D28_['Collected Amount'], data=D28_).set_title("amount collected property wise")
#mlt.savefig('status_20180701_to_20180723.png')
mlt.show()
```

```python
mlt.subplots(figsize=(30,9))
sns.countplot(x=D28_['Property Name'],hue=D28_['Status'], data=D28_).set_title("status Property wise")
#mlt.savefig('status_20180701_to_20180723.png')
mlt.show()
```

```python
mlt.subplots(figsize=(40,7))
sns.countplot(x=D28_['Tenant Code'],hue=D28_['Mode'], data=D28_).set_title("Payment Mode used buy tenant")
#mlt.savefig('status_20180701_to_20180723.png')
mlt.show()
```

```python
mlt.subplots(figsize=(40,7))
sns.countplot(x=D28_['Tenant Code'],hue=D28_['Term'], data=D28_).set_title("Tenant term used")
#mlt.savefig('status_20180701_to_20180723.png')
mlt.show()
```
